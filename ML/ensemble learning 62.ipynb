```python
from sklearn import datasets
```

```python
data = datasets.load_iris()
```

```python
x1 = pd.DataFrame(data.data,columns=["SepelLenght","SepelWidhth"," PetalLenght","PetalWidth"])
```

```python
data.target
```

```python
y1 = pd.DataFrame(data.target,columns=['Species'])
```

```python
data = pd.concat([x1,y1],axis=1)
```

```python
import seaborn as sns
sns.pairplot(data,hue="Species")
```

Here we can see that the 0 rth class which is in yellow this is not mix with any other data point so we are deleting that <br>
we are going to create the classifier on the two classes only 1/2 which are mixed with each other 

```python
op_data = data[data["Species"]!=0][["SepelLenght","SepelWidhth","Species"]]
```

```python
sns.pairplot(op_data,hue="Species")
```

```python
x = op_data.iloc[:,0:2]
```

```python
y = op_data.iloc[:,-1]
```

```python
x
```

```python
y
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.neighbors import KNeighborsClassifier
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import cross_val_score
```

```python
clf1 = LogisticRegression()
clf2 = DecisionTreeClassifier()
clf3 = KNeighborsClassifier()
```

```python
estimators = [('LOG',clf1),('DT',clf2),('KNN',clf3)]
```

```python
print(estimators[0])
print(estimators[1])
print(estimators[2])
```

```python
for estimator in estimators:
    X = cross_val_score(estimator[1],x,y,cv=10,scoring='accuracy')
    print(estimator[0],np.round(np.mean(X),2))
```

# Hard Voting

```python
from sklearn.ensemble import VotingClassifier
```

```python
vc = VotingClassifier(estimators=estimators,voting='hard')
```

```python
vca = cross_val_score(vc,x,y,cv=10,scoring='accuracy')
```

```python
np.round(np.mean(vca),2)
```
