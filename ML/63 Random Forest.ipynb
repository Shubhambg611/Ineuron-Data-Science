```python
from sklearn.svm import SVC
from sklearn.metrics import accuracy_score
```

```python
df = pd.read_csv("D:\Ineuron Data Science\Data\heart.csv")
```

```python
x = df.iloc[:,:-1]
y = df.iloc[:,-1]
```

```python
x
```

```python
y
```

```python
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.25,random_state=42)
```

```python
x_train
```

```python
y_test
```

```python
rf = RandomForestClassifier()
gb = GradientBoostingClassifier()
svc = SVC()
lr = LogisticRegression()
```

```python
rf.fit(x_train,y_train)
y_pred = rf.predict(x_test)
accuracy_score(y_test,y_pred)
```

```python
gb.fit(x_train,y_train)
y_pred2 = gb.predict(x_test)
accuracy_score(y_test,y_pred2)
```

```python
svc.fit(x_train,y_train)
y_pred3 = svc.predict(x_test)
accuracy_score(y_test,y_pred3)
```

```python
lr.fit(x_train,y_train)
y_pred4 = lr.predict(x_test)
accuracy_score(y_test,y_pred4)
```

# hyperparameter tuning

#### Grid search cv


```python
# Number of tress in random forest 
n_estimators = [20,60,100,120]

# number of feature to consider at every split
max_features = [0.2,0.6,1.0]

# Max number of level in tree
max_depth = [2,8,None]

# number of sample 
max_samples = [0.5,0.75,1.0]
```

```python
param_grid = {
    'n_estimators': n_estimators,
    'max_features': max_features,
    'max_depth': max_depth,
    'max_samples': max_samples
}
```

```python
rf3 = RandomForestClassifier()
```

```python
from sklearn.model_selection import GridSearchCV
```

```python
rf3_grid = GridSearchCV(estimator=rf3,param_grid= param_grid,
                       cv=5,
                       verbose=2,
                       n_jobs=-1)

rf3_grid.fit(x_train,y_train)
```

```python
rf3_grid.best_params_
```

```python
rf3_grid.best_score_
```

```python
rf7 = RandomForestClassifier(max_depth= None, max_features=1.0, max_samples =1.0, n_estimators=120)
```

```python

rf7.fit(x_train,y_train)
y_pred7 = rf7.predict(x_test)
accuracy_score(y_test,y_pred7)
```

# random search cv

```python
n_estimators = [20,60,100,120]
max_features = [0.2,0.6,1.0]
max_depth = [2,8,None]
max_samples = [0.5,0.75,1.0]
bootstrap =  [True,False]
min_samples_split = [2,5]
min_samples_leaf = [1,2]
```

```python
param_grid = {'n_estimators': n_estimators,
              'max_features': max_features,
              'max_samples': max_samples,
              'max_depth': max_depth,
              'bootstrap':bootstrap,
              'min_samples_split': min_samples_split,
              'min_samples_leaf':min_samples_leaf

}
print(param_grid)
```

```python
rf9=RandomForestClassifier()
```

```python
from sklearn.model_selection import RandomizedSearchCV

rf9_grid = RandomizedSearchCV(estimator=rf9,
                             param_distributions=param_grid,
                             cv=5,
                             verbose=2,
                             n_jobs=-1)
```

```python
rf9_grid.fit(x_train,y_train)
```

```python
rf9_grid.best_params_
```

```python
rf10=RandomForestClassifier(n_estimators=60, min_samples_split=5, min_samples_leaf=1, max_samples=0.75,max_features=0.6, max_depth=8, bootstrap= True)
```

```python
rf10.fit(x_train,y_train)
y_pred10 = rf10.predict(x_test)
accuracy_score(y_test,y_pred10)
```
