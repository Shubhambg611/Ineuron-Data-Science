# SVM kernels and SVM regression

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.svm import SVC
from mpl_toolkits import mplot3d
from matplotlib.axes._axes import _log as matplotlib_axes_logger
from matplotlib.colors import ListedColormap
```

### Samaple Data generation (using sklearn sample generator dataset)

```python
from sklearn.datasets import make_circles
```

```python
x,y = make_circles(100,factor=0.1,noise=0.1)
```

```python
x
```

```python
y #lable of the data 
```

```python
plt.scatter(x[:,0],x[:,1],c=y,s=50,cmap='bwr')
```

```python
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.20)
```

```python
x_train
```

For svc we have used linear hear which is traditional method which is not capable of classifiying the 2d data 

### Linear

```python
classifier = SVC(kernel="linear")
```

```python
classifier.fit(x_train,y_train)
```

```python
y_pred = classifier.predict(x_test)
```

```python
y_pred
```

```python
# actual values
y_test
```

```python
from sklearn.metrics import accuracy_score
```

```python
accuracy_score(y_test,y_pred)*100
```

Now using diff kernels to increase the accuracy

### RBF

```python
classifier2=SVC(kernel="rbf")
```

```python
classifier2.fit(x_train,y_train)
```

```python
y_pred2 = classifier2.predict(x_test)
```

```python
y_test
```

```python
y_pred2
```

```python
accuracy_score(y_test,y_pred2)*100
```

Kernel transforming data intyo higher dimension

### Polynomial

```python
classifier3 = SVC(kernel="poly")
```

```python
classifier3.fit(x_train,y_train)
```

```python
y_pred3 = classifier3.predict(x_test)
```

```python
accuracy_score(y_test,y_pred3)*100
```

Now there is one hyper parameter for the polynomial which is degree will try to make chnages with respect to the degree value and cheack will it improv ethe accuracy o fhe model

```python
classifier3_1 = SVC(kernel="poly",degree=2)
```

```python
classifier3_1.fit(x_train,y_train)
```

```python
y_pred3_1 = classifier3_1.predict(x_test)
```

```python
accuracy_score(y_test,y_pred3_1)*100
```

### Sigmoid

```python
classifier4 = SVC(kernel="sigmoid")
```

```python
classifier4.fit(x_train,y_train)
```

```python
y_pred4 = classifier4.predict(x_test)
```

```python
accuracy_score(y_test,y_pred4)*100
```
